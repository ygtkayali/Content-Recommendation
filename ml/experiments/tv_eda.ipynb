```python
tv_df = pd.read_csv("../../data/raw/tvs.csv")
```

```python
# combine origin_country[0]...origin_country[7] into one column. If lets say origin_countr[1] is null then skip it, if exists add it to a list and then join the list with a comma and add it to a new column called origin_countries. Do this for all rows. After that drop the origin_country[0]...origin_country[7] columns
origin_country_cols = [f"origin_country[{i}]" for i in range(8)]
def combine_origin_countries(row):
    countries = []
    for col in origin_country_cols:
        if pd.notnull(row[col]):
            countries.append(row[col])
    return ", ".join(countries)
tv_df["origin_countries"] = tv_df.apply(combine_origin_countries, axis=1)
tv_df.drop(columns=origin_country_cols, inplace=True)
```

```python
#combine genres[0].name...genres[7].name into one column called genres. If lets say genres[1].name is null then skip it, if exists add it to a list and then join the list with a comma and add it to a new column called genres. Do this for all rows. After that drop the genres[0].name...genres[7].name columns
genre_cols = [f"genres[{i}].name" for i in range(8)]
def combine_genres(row):
    genres = []
    for col in genre_cols:
        if pd.notnull(row[col]):
            genres.append(row[col])
    return ", ".join(genres)
tv_df["genres"] = tv_df.apply(combine_genres, axis=1)
tv_df.drop(columns=genre_cols, inplace=True)
```
